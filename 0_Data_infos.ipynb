# Importation des données

```python
import glob # Permet de rechercher tous les chemins ayant le motif spécifié
import numpy as np
import pandas as pd

import cv2 # Open CV
```

```python
# On recherche tous les .jpg dans les dossiers de niveau +2 (données d'entrainement et test)
path_images_train_test = glob.glob(pathname='*/'+ '*/'+ '*/' + '*/'+ '*.jpg') 
```

```python
# On recherche tous les JPG dans les dossiers de niveau +1 (données de prédiction)
path_images_pred = glob.glob(pathname='*/'+ '*/'+ '*/' + '*.jpg') 
```

# Extraction d'informations des données

```python
# On affiche le nombre d'images trouvées pour chaque jeu de données
print(f"data_train_&_test: {len(path_images_train_test)}")

print(f"data_pred: {len(path_images_pred)}")
```

```python
# On crée un dataframe en fonction de la liste "path_images_train_test"
data_train_test = pd.DataFrame({'path_image':path_images_train_test}) 

# On crée un dataframe en fonction de la liste "path_images_pred"
data_pred = pd.DataFrame({'path_image':path_images_pred}) 
```

```python
print("Jeu de données Entrainement et Test")
display(data_train_test)
print("\n\n")
print("Jeu de données Prédiction")
display(data_pred)
```

# Création de la variable "set"

```python
'''On crée une variable 'set' en fonction de la variable "path_images_train_test", 
et on recupère le nom du set en utilisant split'''
data_train_test['set'] = data_train_test['path_image'].apply(lambda x: x.split('/')[2])


'''On crée une variable 'set' en fonction de la variable "path_images_pred", 
et on recupère le nom du set en utilisant split'''
data_pred['set'] = data_pred['path_image'].apply(lambda x: x.split('/')[2])
```

# Création de la variable "category"

```python
'''On crée une variable 'category' en fonction de la variable "path_images_train_test",
et on recupère le nom du la catégorie en utilisant split'''
data_train_test['category'] = data_train_test['path_image'].apply(lambda x: x.split('/')[3])


'''On crée une variable 'category' en fonction de la variable "path_images_pred",
et on ne met rien, la catégorie des données de prédiction étant la variable à discriminer'''
data_pred['category'] = ""
```

# Création de la variable "name"

```python
'''On crée une variable 'name' en fonction de la variable "path_images_train_test",
et on recupère le nom d'image en utilisant split'''
data_train_test['name'] = data_train_test['path_image'].apply(lambda x: x.split('/')[-1].split('.')[0])


'''On crée une variable 'name' en fonction de la variable "path_images_pred",
et on recupère le nom d'image en utilisant split'''
data_pred['name'] = data_pred['path_image'].apply(lambda x: x.split('/')[-1].split('.')[0])
```

# Création de la variable "extension"

```python
'''On crée une variable 'extension' en fonction de la variable "path_images_train_test",
et on recupère l'extention en utilisant split'''
data_train_test['extension'] = data_train_test['path_image'].apply(lambda x: x.split('.')[-1])


'''On crée une variable 'extension' en fonction de la variable "path_images_pred",
et on recupère l'extention en utilisant split'''
data_pred['extension'] = data_pred['path_image'].apply(lambda x: x.split('.')[-1])
```

# Récupération des information de dimension des données (width, height, size)

```python
'''On crée 3 listes pour récuperer les différentes informations de 
la taille des images (hauteur, largeur et nombre de dimension)'''
img_width_train_test, img_height_train_test, img_dim_train_test,  img_size_train_test = [],[],[],[]

'''Pour chaque image dans 'path_images_train_test',
on lit l'image puis on stocke les infos dans chaque liste respectives'''
for image in path_images_train_test:
    img = cv2.imread(image, cv2.IMREAD_COLOR)
    img_width_train_test.append(img.shape[0])
    img_height_train_test.append(img.shape[1])
    img_dim_train_test.append(img.shape[2])
    img_size_train_test.append(str(img.shape[0]) + ' x ' + str(img.shape[1]))
    
    
'''On crée 3 listes pour récuperer les différentes informations de 
la taille des images (hauteur, largeur et nombre de dimension)'''
img_width_pred, img_height_pred, img_dim_pred, img_size_pred = [],[],[], []

'''Pour chaque image dans 'path_images_pred',
on lit l'image puis on stocke les infos dans chaque liste respectives'''
for image in path_images_pred:
    img = cv2.imread(image, cv2.IMREAD_COLOR)
    img_width_pred.append(img.shape[0])
    img_height_pred.append(img.shape[1])
    img_dim_pred.append(img.shape[2])

    img_size_pred.append(str(img.shape[0]) + ' x ' + str(img.shape[1])) 
```

# On regroupe respectivement les dataframes "data_train_test" et "data_pred" avec les informations de dimension

```python
'''On regroupe le dataframe 'data_train_test' avec un nouveau dataframe contenant les informations 
des listes créees précedement'''
data_train_test = pd.concat([data_train_test,
                            pd.DataFrame({'width':img_width_train_test, 
                                          'height':img_height_train_test, 
                                          'dim':img_dim_train_test,
                                          'size':img_size_train_test})], axis=1)


'''On regroupe le dataframe 'data_pred' avec un nouveau dataframe contenant les informations 
des listes créees précedement'''
data_pred = pd.concat([data_pred,
                       pd.DataFrame({'width':img_width_pred,
                                     'height':img_height_pred,
                                     'dim':img_dim_pred,
                                     'size':img_size_pred})], axis=1)
```

```python
print("Jeu de données Entrainement et Test")
display(data_train_test)
print("\n\n")
print("Jeu de données de Prédiction")
display(data_pred)
```

# SAUVEGARDE DES DONNEES

```python
data_train_test.to_csv("data_train_test", index=False)

data_pred.to_csv("data_pred", index=False)
```
